## 1. Categorical Data

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
sns.countplot(x=df['Survived'])
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x='Pclass',data=df)
```

```python
sns.countplot(x=df['Sex'])
```

```python
df['Embarked'].value_counts().plot(kind='barh',xlabel='Count',color='green')
```

```python
df['Survived'].value_counts().plot(kind='pie',title='Percentage of Servived',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',title='percentage of Pclass',autopct='%.2f')
```

```python
df['Sex'].value_counts().plot(kind='pie',title='Percentage of Gender', autopct='%.2f')
```
