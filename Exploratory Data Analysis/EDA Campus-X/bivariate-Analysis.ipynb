#### Scatter plot multivariate analysis (Numerical - Numerical) ####

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a Seaborn style
sns.set(style="whitegrid", palette="muted")

# Create the scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips, 
                hue='sex', style='smoker', size='size', 
                sizes=(40, 200),  markers=['o', 'X'])

# Customize the title and labels
plt.title('Scatterplot of Total Bill vs Tip', fontsize=16, weight='bold')
plt.xlabel('Total Bill ($)', fontsize=12, weight='bold')
plt.ylabel('Tip ($)', fontsize=12, weight='bold')

# Adjust the legend
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='best')

# Show the plot
plt.tight_layout()
plt.show()
```

## Distplot (Numerical - Categorical) ##

```python

sns.displot(titanic[titanic['Survived']==0]['Age'],kde=True)
plt.title('Not Survived')
sns.displot(titanic[titanic['Survived']==1]['Age'],kde=True)
plt.title('Survived')
```

```python
sns.displot(data=titanic, x='Age', hue='Survived', kind='hist', fill=True)

# Customize the title and labels
plt.title('Age Distribution of Survived vs Not Survived', fontsize=16, weight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sns.displot(data=titanic, x='Age', hue='Survived', kind='kde',fill=True)

# Customize the title and labels
plt.title('Age Distribution of Survived vs Not Survived', fontsize=16, weight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
